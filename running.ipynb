```python
# Import the Pandas library
import pandas as pd
from datetime import datetime
dateparse = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
import requests

import warnings
warnings.filterwarnings('ignore')

# Import API key
from api_keys import weather_api_key
```

```python
# Create a reference to the CSV file desired
csv_path = "data/activities-garmin.csv"

# Read the CSV into a Pandas DataFrame
garmin_data_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
garmin_data_df.head()
```

```python
#drop all non-running activities'

cleaning_running_df = garmin_data_df.loc[(
    garmin_data_df["Activity Type"] == "Run")]

cleaning_running_df.head()
```

```python
#Create a dataframe using only hte columns we want

running_df = cleaning_running_df[["Activity Date","Activity Name", "Activity Type", "Elapsed Time", "Distance", "Max Heart Rate", "Moving Time", "Average Heart Rate"]]

running_df.head(200)
```

```python
#check the data types 
running_df.dtypes
```

```python
#dropping rows with missing data

run_df = running_df.dropna(how='any')

clean_running_df = run_df.loc[(
    run_df["Distance"] > 2.00)]

clean_running_df
```

```python
#Convert the date to a time stamp

from datetime import datetime

clean_running_df["Time_Stamp"]=pd.to_datetime(clean_running_df["Activity Date"]).map(pd.Timestamp.timestamp)
clean_running_df.head()
```

```python
#convert the Activit date to a datetime variable
clean_running_df["Activity Date"]=pd.to_datetime(clean_running_df["Activity Date"])
clean_running_df.head()
```

```python
#check data types again

clean_running_df.dtypes
```

```python
#Create a smaller subset of data to eliminate the need to buy API calls.

running_test_df = clean_running_df.head(10)

running_test_df
```

```python
#api call

#declare a dictionary for API data
running_weather_data = []

#Create a list of datestamps for API Calls
ds_list = running_test_df["Time_Stamp"].astype("str").to_list()
ds_list = [sub[: -2] for sub in ds_list] 

#Generate the API url.
url = "https://api.openweathermap.org/data/3.0/onecall/timemachine?lat=39.952583&lon=-75.165222&dt=" 
api_key = "&appid=" + weather_api_key

#interative loop to for API calls.
for ts in ds_list:
    
#API calls
    weather_url = url + ts + api_key
    weather = requests.get(weather_url).json()
    
    #pull out the data we want from the json
    
    temp_js = weather["data"][0]["temp"]
    humidity_js = weather["data"][0]["humidity"]
    wind_speed_js = weather["data"][0]["wind_speed"]
    
    #convert kelvin temperature to farhenheit
    temp_f = round(1.8*(temp_js-273) + 32,1)
        
    #create the dictionary    
    running_weather_data.append({
        
        "timestamp" : ts,
        "temp" : temp_f,
        "humidity" : humidity_js,
        "wind_speed" : wind_speed_js               
    })
```

```python
#convert the dictionary to a dataframe

running_weather_df=pd.DataFrame(running_weather_data)
running_weather_df.head(10)
```

```python
#checking datatypes on running_weather_df
running_weather_df.dtypes
```

```python
#convert running_weather_df.timestamp to a float.
running_weather_df.timestamp = running_weather_df.timestamp.astype('float')

#join running_test_df with running_weather_df
running_test_df.set_index('Time_Stamp').join(running_weather_df.set_index('timestamp'))
```
